```python
# åŠ è½½æ•°æ®
def opencsv():
    # ä½¿ç”¨ pandas æ‰“å¼€
    data = pd.read_csv('train.csv')
    data1 = pd.read_csv('test.csv')

    train_data = data.values[0:, 1:]  # è¯»å…¥å…¨éƒ¨è®­ç»ƒæ•°æ®,  [è¡Œï¼Œåˆ—]
    train_label = data.values[0:, 0] # è¯»å–åˆ—è¡¨çš„ç¬¬ä¸€åˆ—
    test_data = data1.values[0:, 0:]  # æµ‹è¯•å…¨éƒ¨æµ‹è¯•ä¸ªæ•°æ®
    return train_data, train_label, test_data
```

```python
# åŠ è½½æ•°æ®
trainData, trainLabel, testData = opencsv()
print("trainData==>", type(trainData), shape(trainData))
print("trainLabel==>", type(trainLabel), shape(trainLabel))
print("testData==>", type(testData), shape(testData))
```

```python
def knnClassify(trainData, trainLabel):
    knnClf = KNeighborsClassifier()   # default:k = 5,defined by yourself:KNeighborsClassifier(n_neighbors=10)
    knnClf.fit(trainData, ravel(trainLabel)) # ravel Return a contiguous flattened array.
    return knnClf
```

```python
start_time = time.time()#çœ‹çœ‹è®­ç»ƒæ¨¡åž‹éœ€è¦å¤šä¹…
knnClf = KNeighborsClassifier()   # default:k = 5,defined by yourself:KNeighborsClassifier(n_neighbors=10)
knnClf.fit(trainData, ravel(trainLabel)) # ravel Return a contiguous flattened array.
print("load data finish")
stop_time_l = time.time()
print('KNN model time used:%f' % (stop_time_l - start_time))
```

```python
start_time = time.time()#çœ‹çœ‹ç”¨è®­ç»ƒå¥½çš„æ¨¡åž‹æ¥é¢„æµ‹testlabeléœ€è¦å¤šä¹…
# ç»“æžœé¢„æµ‹
testLabel = knnClf.predict(testData)
print("predict finish")
stop_time_l = time.time()
print('Prediction time:%f' % (stop_time_l - start_time))
```

# ä¸‹é¢åœ¨å°†data é™ç»´ä¹‹åŽçœ‹çœ‹é¢„æµ‹ç»“æžœ #

```python
# æ•°æ®é¢„å¤„ç†-é™ç»´ PCAä¸»æˆæˆåˆ†åˆ†æž
import numpy as np
from sklearn.decomposition import PCA
def dRPCA(x_train, x_test, COMPONENT_NUM):
    print('dimensionality reduction...')
    trainData = np.array(x_train)
    testData = np.array(x_test)
    '''
    ä½¿ç”¨è¯´æ˜Žï¼šhttps://www.cnblogs.com/pinard/p/6243025.html
    n_components>=1       n_components=NUM  ^H è®¾ç½®^Hå ç‰¹å¾æ•°é‡æ¯”
    0 < n_components < 1  n_components=0.99  ^Hè®¾ç½®é˜ˆå€¼æ€»æ–¹å·®å æ¯”
    '''
    pca = PCA(n_components=COMPONENT_NUM, whiten=True)
    # whiten ï¼šåˆ¤æ–­æ˜¯å¦è¿›è¡Œç™½åŒ–ã€‚æ‰€è°“ç™½åŒ–ï¼Œå°±æ˜¯å¯¹é™ç»´åŽçš„æ•°æ®çš„æ¯ä¸ªç‰¹å¾è¿›è¡Œå½’ä¸€åŒ–ï¼Œ
    # è®©æ–¹å·®éƒ½ä¸º1.å¯¹äºŽPCAé™ç»´æœ¬èº«æ¥è¯´ï¼Œä¸€èˆ¬ä¸éœ€è¦ç™½åŒ–ã€‚å¦‚æžœä½ PCAé™ç»´åŽæœ‰åŽç»­çš„æ•°æ®å¤„ç†åŠ¨ä½œï¼Œ
    # å¯ä»¥è€ƒè™‘ç™½åŒ–ã€‚é»˜è®¤å€¼æ˜¯Falseï¼Œå³ä¸è¿›è¡Œç™½åŒ–ã€‚
    pca.fit(trainData)  # Fit the model with X
    pcaTrainData = pca.transform(trainData)  # Fit the model with X and åœ¨Xä¸Šå®Œæˆé™ç»´.
    pcaTestData = pca.transform(testData)  # Fit the model with X and åœ¨Xä¸Šå®Œæˆé™ç»´.

    # pca æ–¹å·®å¤§å°ã€æ–¹å·®å æ¯”ã€ç‰¹å¾æ•°é‡
    print(pca.explained_variance_, '\n', pca.explained_variance_ratio_, '\n',
          pca.n_components_)
    # explained_variance_:å®ƒä»£è¡¨é™ç»´åŽçš„å„ä¸»æˆåˆ†çš„æ–¹å·®å€¼ã€‚æ–¹å·®å€¼è¶Šå¤§ï¼Œåˆ™è¯´æ˜Žè¶Šæ˜¯é‡è¦çš„ä¸»æˆåˆ†ã€‚
    # explained_variance_ratio_ï¼Œå®ƒä»£è¡¨é™ç»´åŽçš„å„ä¸»æˆåˆ†çš„æ–¹å·®å€¼å æ€»æ–¹å·®å€¼çš„æ¯”ä¾‹ï¼Œè¿™ä¸ªæ¯”ä¾‹è¶Šå¤§ï¼Œåˆ™è¶Šæ˜¯é‡è¦çš„ä¸»æˆåˆ†ã€‚
    print(sum(pca.explained_variance_ratio_))
    return pcaTrainData, pcaTestData
```

```python
 # é™ç»´å¤„ç†
trainData, testData = dRPCA(trainData, testData, 35)
```

```python
# æ¨¡åž‹è®­ç»ƒ
knnClf = knnClassify(trainData, trainLabel)

```

```python
# ç»“æžœé¢„æµ‹
start_time = time.time()#çœ‹çœ‹ç”¨è®­ç»ƒå¥½çš„æ¨¡åž‹æ¥é¢„æµ‹testlabeléœ€è¦å¤šä¹…
# ç»“æžœé¢„æµ‹
testLabel = knnClf.predict(testData)
print("predict finish")
stop_time_l = time.time()
print('Prediction time:%f' % (stop_time_l - start_time))
```
