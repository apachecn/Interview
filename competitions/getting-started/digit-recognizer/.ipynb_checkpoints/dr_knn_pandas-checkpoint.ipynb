每一行数据是： 目标变量+特征变量值。它是将图片的像素数据转化而成的文本数据

```python
# 加载数据
def opencsv():
    # 使用 pandas 打开
    data = pd.read_csv('train.csv')
    data1 = pd.read_csv('test.csv')

    train_data = data.values[0:, 1:]  # 读入全部训练数据,  [行，列]
    train_label = data.values[0:, 0] # 读取列表的第一列
    test_data = data1.values[0:, 0:]  # 测试全部测试个数据
    return train_data, train_label, test_data
```

```python
# 加载数据
trainData, trainLabel, testData = opencsv()
print("trainData==>", type(trainData), shape(trainData))
print("trainLabel==>", type(trainLabel), shape(trainLabel))
print("testData==>", type(testData), shape(testData))
```

```python
def knnClassify(trainData, trainLabel):
    knnClf = KNeighborsClassifier()   # default:k = 5,defined by yourself:KNeighborsClassifier(n_neighbors=10)
    knnClf.fit(trainData, ravel(trainLabel)) # ravel Return a contiguous flattened array.
    return knnClf
```

```python
start_time = time.time()#看看训练模型需要多久
knnClf = KNeighborsClassifier()   # default:k = 5,defined by yourself:KNeighborsClassifier(n_neighbors=10)
knnClf.fit(trainData, ravel(trainLabel)) # ravel Return a contiguous flattened array.
print("load data finish")
stop_time_l = time.time()
print('KNN model time used:%f' % (stop_time_l - start_time))
```

```python
start_time = time.time()#看看用训练好的模型来预测testlabel需要多久
# 结果预测
testLabel = knnClf.predict(testData)
print("predict finish")
stop_time_l = time.time()
print('Prediction time:%f' % (stop_time_l - start_time))
```

预测竟然用了半个小时

```python
def saveResult(result, csvName):
    with open(csvName, 'wb') as myFile: # 创建记录输出结果的文件（w 和 wb 使用的时候有问题）
    #python3里面对 str和bytes类型做了严格的区分，不像python2里面某些函数里可以混用。所以用python3来写wirterow时，打开文件不要用wb模式，只需要使用w模式，然后带上newline=''
        myWriter = csv.writer(myFile) # 对文件执行写入
        myWriter.writerow(["ImageId", "Label"]) # 设置表格的列名
        index = 0
        for i in result:
            tmp = []
            index = index + 1
            tmp.append(index)
            # tmp.append(i)
            tmp.append(int(i)) # 测试集的标签值
            myWriter.writerow(tmp)
```

# 下面在将data 降维之后看看预测结果 #

```python
# 数据预处理-降维 PCA主成成分分析
import numpy as np
from sklearn.decomposition import PCA
def dRPCA(x_train, x_test, COMPONENT_NUM):
    print('dimensionality reduction...')
    trainData = np.array(x_train)
    testData = np.array(x_test)
    '''
    使用说明：https://www.cnblogs.com/pinard/p/6243025.html
    n_components>=1       n_components=NUM  ^H 设置^H占特征数量比
    0 < n_components < 1  n_components=0.99  ^H设置阈值总方差占比
    '''
    pca = PCA(n_components=COMPONENT_NUM, whiten=True)
    # whiten ：判断是否进行白化。所谓白化，就是对降维后的数据的每个特征进行归一化，
    # 让方差都为1.对于PCA降维本身来说，一般不需要白化。如果你PCA降维后有后续的数据处理动作，
    # 可以考虑白化。默认值是False，即不进行白化。
    pca.fit(trainData)  # Fit the model with X
    pcaTrainData = pca.transform(trainData)  # Fit the model with X and 在X上完成降维.
    pcaTestData = pca.transform(testData)  # Fit the model with X and 在X上完成降维.

    # pca 方差大小、方差占比、特征数量
    print(pca.explained_variance_, '\n', pca.explained_variance_ratio_, '\n',
          pca.n_components_)
    # explained_variance_:它代表降维后的各主成分的方差值。方差值越大，则说明越是重要的主成分。
    # explained_variance_ratio_，它代表降维后的各主成分的方差值占总方差值的比例，这个比例越大，则越是重要的主成分。
    print(sum(pca.explained_variance_ratio_))
    return pcaTrainData, pcaTestData
```

```python
 # 降维处理
trainData, testData = dRPCA(trainData, testData, 35)
```

```python
# 模型训练
knnClf = knnClassify(trainData, trainLabel)

```

```python
# 结果预测
start_time = time.time()#看看用训练好的模型来预测testlabel需要多久
# 结果预测
testLabel = knnClf.predict(testData)
print("predict finish")
stop_time_l = time.time()
print('Prediction time:%f' % (stop_time_l - start_time))
```

** 经过pca降维后预测时间果然少了好多，降维后用到的时间差不多是降维前的十分之一吧 **

```python
# 结果的输出
saveResult(testLabel,'Result_sklearn_knn_pca.csv')
print("finish!")
```
